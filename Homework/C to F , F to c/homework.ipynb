```python
# โปรแกรมแปลง Celsius ↔ Fahrenheit แบบอัตโนมัติ

temp = input("ป้อนอุณหภูมิ (เช่น 30C หรือ 86F): ")

# แยกค่าตัวเลขและตัวอักษร
value = float(temp[:-1]) #ตัวเลข
unit = temp[-1].upper() #ตัวอักษร C,F

if unit == "C":
    result = (value * 9/5) + 32
    print(f"{value}°C = {result}°F")

elif unit == "F":
    result = (value - 32) * 5/9
    print(f"{value}°F = {result}°C")

else:
    print("รูปแบบข้อมูลไม่ถูกต้อง! กรุณาใส่ข้อมูลตามนี้ เช่น 30C หรือ 86F")
    #นายปฐมพงศ์ พึ่งประโคน 684245008
    #นายพัศณพล คานภู่     684245010
    #นาย กฤตพิภัช ศรีนิล    684245002
```
