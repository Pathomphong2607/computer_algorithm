```python
import random
a = ['Drink','food','Gereral','Cloth']
list1 = []
for i in range(random.randint(10,15)):
    id=i+1
    name = "Product"+id.__str__()
    category = random.choice(a)
    price = random.randint(10,30)
    list1.append((id,name,category,price))
for item in list1:
    print(item)


```

```python
findhighest = max(list1, key=lambda x: x[3])
for item in list1:
    if item[3] == findhighest[3]:
        print(item)
```

```python
findlowest = min(list1, key=lambda x: x[3])
for item in list1:
    if item[3] == findlowest[3]:
        print(item)
```

```python
for Cat in a:
    count=0
    item_cat=[]
    for item in list1:
        if item[2]==Cat:
            count=+1
            item_cat.append(item)
            print(f"\n{Cat}{count}")
            for item in item_cat:
                print(item)
```
